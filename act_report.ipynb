# Analysis of WeRateDogs Twitter Page

WeRateDogs is a twitter page which rates pictures of dogs in a humorous way. In this analysis, the dataset used contains over 2000 of the tweets from their page, containing information about the retweet count, favorite count, tweet ID, as well as the rating which the author provided to each dog, as well as predictions on the species of dog which appears in the photo, made by a neural network. 

## Bivariate Insights

As can be seen in this graph, the ratings numerator provided by the user is somewhat predictive of the total favorite count. As the numerator increases, so too does the favorite count of that tweet. 

As can be seen in this graph, the ratings numerator provided by the user is also somewhat predictive of the total retweet count. As the numerator increases, so too does the retweet count of that tweet. 

## Time Series Observations

Over time, the number of favorites per tweet has increased. This is unsurprising, as the twitter page has increased its following over the past several years, and thus its more recent posts should be expected to receive a higher number of favorites on average. 

Interestingly, the rating numerator, which is provided by the account, has also increased over time. It appears that the author of the page has become more lenient over time, and is now much more likely to give ratings that are greater than 10, and very unlikely to give any rating below that. 

## Zipf's Law and Pareto Principle

- Total Images = 2067
- Dog breeds in the top 20% = 76
- Images of breeds in the top 20% = 1494
- Proportion of images of top 20% breeds to total images = 72%

There appears to be an inverse relationship between the frequency of a dog breed's appearance, and the frequency rank of that particular breed. When you set both axes to log, then an almost linear relationship appears. This is a manifestation of zipf's law, which, in the application of linguistics, that the frequency of a word is inversely proportional to its rank in the frequency table. This is a universal phenomenon that has cropped up in a plethora of different fields, and thus it is not suprising that we see it here. 

This is also evidence of the pareto principle: that for many events, roughly 80% of the effects come from 20% of the causes. Based on the calculations, the top 20% most frequently occuring dog breeds (76 breeds), accounts for about 72% of all the dog images (1494 / 2067), which is not far off from the 80% predicted by the principle. Perhaps with a larger sample size, the results may be different.
